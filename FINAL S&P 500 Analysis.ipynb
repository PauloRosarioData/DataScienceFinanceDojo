```python
df = pd.read_csv('/home/nbuser/library/all_stocks_5yr.csv')
del df['Unnamed: 7']
del df['Unnamed: 8']
del df['Unnamed: 9']
del df['Unnamed: 10']
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df
```

```python
#Strategy 1 - invest in companies beginning with the letter A
investment = 1000000
strategy_1_returns = list()
for i in range(2013, 2019):
    df2 = df.loc[df['Date'].dt.year == i]
    names = df2['Name'].drop_duplicates()
    companies_letter_A = list()
    for name in names:
        if name[0] == 'A':
            companies_letter_A.append(name)
    table1 = pd.DataFrame(companies_letter_A)
    table1 = table1.rename(columns = {0:'Company Name'})
    share_investment = investment/len(companies_letter_A)
    start_share_price = list()
    for name in companies_letter_A:
        min_date = df2.query('Name == "{}"'.format(name))['Date'].min()
        start_share_price.append(df2['Close'][(df2['Date'] == min_date)&(df2['Name'] == name)].item())
    table1['Start Share Price'] = start_share_price
    table1['Stock'] = share_investment/table1['Start Share Price']
    end_share_price = list()
    for name in companies_letter_A:
        max_date = df2.query('Name == "{}"'.format(name))['Date'].max()
        end_share_price.append(df2['Close'][(df2['Date'] == max_date)&(df2['Name'] == name)].item())
    table1['End Share Price'] = end_share_price
    table1['Return'] = table1['Stock']*table1['End Share Price']
    total_return = table1['Return'].sum()
    print(i)
    print(total_return)
    strategy_1_returns.append(total_return)
    investment = total_return
```

```python
#Strategy 2 - invest in top 25 companies from previous year (Momentum strategy)
investment = 1000000
strategy_2_returns = list()
for i in range(2013, 2018):
    #Finding the top 25 companies from previous year
    df3 = df.loc[df['Date'].dt.year == i]
    companies = df3['Name'].drop_duplicates().tolist()
    table2 = pd.DataFrame(companies)
    table2 = table2.rename(columns = {0: 'Company Name'})
    start_share_price_previous_year = list()
    end_share_price_previous_year = list()
    for name in companies:
        min_date = df3.query('Name == "{}"'.format(name))['Date'].min()
        start_share_price_previous_year.append(df3['Close'][(df3['Date'] == min_date)&(df3['Name'] == name)].item())  
        max_date = df3.query('Name == "{}"'.format(name))['Date'].max()
        end_share_price_previous_year.append(df3['Close'][(df3['Date'] == max_date)&(df3['Name'] == name)].item())
    table2['Start Share Price'] = start_share_price_previous_year
    table2['End Share Price'] = end_share_price_previous_year
    table2['Change in Share Price'] = table2['End Share Price'] - table2['Start Share Price']
    table2.sort_values('Change in Share Price', ascending=False, inplace=True)
    table3 = table2.head(25)
    #Investing in subsequent year
    df3 = df.loc[df['Date'].dt.year == (i+1)]
    top_25_performers = table3['Company Name'].tolist()
    top_25_performers_df = pd.DataFrame(top_25_performers)
    top_25_performers_df = top_25_performers_df.rename(columns = {0:'Top 25 Companies'})
    start_share_price_next_year = list()
    end_share_price_next_year = list()
    for name in top_25_performers:
        min_date = df3.query('Name == "{}"'.format(name))['Date'].min()
        start_share_price_next_year.append(df3['Close'][(df3['Date'] == min_date)&(df3['Name'] == name)].item())
        max_date = df3.query('Name == "{}"'.format(name))['Date'].max()
        end_share_price_next_year.append(df3['Close'][(df3['Date'] == max_date)&(df3['Name'] == name)].item())
    top_25_performers_df['Start Share Price'] = start_share_price_next_year
    share_investment = investment/25
    top_25_performers_df['Stock'] = share_investment/top_25_performers_df['Start Share Price']
    top_25_performers_df['End Share Price'] = end_share_price_next_year
    top_25_performers_df['Return'] = top_25_performers_df['Stock']*top_25_performers_df['End Share Price']
    total_return_year_after = top_25_performers_df['Return'].sum()
    print((i+1))
    print(total_return_year_after)
    strategy_2_returns.append(total_return_year_after)
    investment = total_return_year_after
```

```python
#Strategy 3 - invest in bottom 25 companies from previous year (Dogs strategy)
investment = 1000000
strategy_3_returns = list()
for i in range(2013, 2018):
    #Finding the bottom 25 companies from previous year 
    df3 = df.loc[df['Date'].dt.year == i]
    companies = df3['Name'].drop_duplicates().tolist()
    table2 = pd.DataFrame(companies)
    table2 = table2.rename(columns = {0: 'Company Name'})
    start_share_price_previous_year = list()
    end_share_price_previous_year = list()
    for name in companies:
        min_date = df3.query('Name == "{}"'.format(name))['Date'].min()
        start_share_price_previous_year.append(df3['Close'][(df3['Date'] == min_date)&(df3['Name'] == name)].item())  
        max_date = df3.query('Name == "{}"'.format(name))['Date'].max()
        end_share_price_previous_year.append(df3['Close'][(df3['Date'] == max_date)&(df3['Name'] == name)].item())
    table2['Start Share Price'] = start_share_price_previous_year
    table2['End Share Price'] = end_share_price_previous_year
    table2['Change in Share Price'] = table2['End Share Price'] - table2['Start Share Price']
    table2.sort_values('Change in Share Price', ascending=False, inplace=True)
    table3 = table2.tail(25)
    #Investing in subsequent year
    df3 = df.loc[df['Date'].dt.year == (i+1)]
    bottom_25_performers = table3['Company Name'].tolist()
    bottom_25_performers_df = pd.DataFrame(bottom_25_performers)
    bottom_25_performers_df = bottom_25_performers_df.rename(columns = {0:'Bottom 25 Companies'})
    start_share_price_next_year = list()
    end_share_price_next_year = list()
    for name in bottom_25_performers:
        min_date = df3.query('Name == "{}"'.format(name))['Date'].min()
        start_share_price_next_year.append(df3['Close'][(df3['Date'] == min_date)&(df3['Name'] == name)].item())
        max_date = df3.query('Name == "{}"'.format(name))['Date'].max()
        end_share_price_next_year.append(df3['Close'][(df3['Date'] == max_date)&(df3['Name'] == name)].item())
    bottom_25_performers_df['Start Share Price'] = start_share_price_next_year
    share_investment = investment/25
    bottom_25_performers_df['Stock'] = share_investment/bottom_25_performers_df['Start Share Price']
    bottom_25_performers_df['End Share Price'] = end_share_price_next_year
    bottom_25_performers_df['Return'] = bottom_25_performers_df['Stock']*bottom_25_performers_df['End Share Price']
    total_return_year_after = bottom_25_performers_df['Return'].sum()
    print((i+1))
    print(total_return_year_after)
    strategy_3_returns.append(total_return_year_after)
    investment = total_return_year_after
```

```python
all_strategies = pd.DataFrame(index=range(2014, 2019), columns=['Strategy 1', 'Strategy 2', 'Strategy 3'])
all_strategies['Strategy 1'] = strategy_1_returns[1:]
all_strategies['Strategy 2'] = strategy_2_returns
all_strategies['Strategy 3'] = strategy_3_returns
all_strategies
```

```python
for strategy in all_strategies.columns:
    x = all_strategies.index
    y = all_strategies[strategy] 
    plt.plot(x, y)
    plt.legend(['Strategy 1', 'Strategy 2', 'Strategy 3'])
plt.xlabel('Year')
plt.ylabel('Total Return')
```
